```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f"User uploaded file: {filename}")
    df = pd.read_csv(filename)

df.head()
```

```python
print("Dataset Info")
df.info()

print("Summary Statistics")
print(df.describe())

print("checking duplicate rows")
print(df.duplicated().sum())

print("checking missing values")
print(df.isnull().sum())
```

```python
# Categorical and Numerical column Analysis
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
print('\nCategorical Columns:')
print(categorical_columns)
print('\nNumerical Columns:')
print(numerical_columns)

print("\nUnique values in Categorical Columns")
for column in categorical_columns:
    unique_values = df[column].unique()
    print(f"\nColumn: {column}")
    print(f"Unique Values: {unique_values}")
```

# Univariate Analysis
Distribution for Numerical Columns

```python
df[numerical_columns].hist(figsize=(12,8) ,bins=30, color='skyblue',edgecolor='black')
plt.suptitle("Distribution of Numerical Features ", fontsize=14)
plt.show()
```

#  Boxplots for Outliers

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numerical_columns])
plt.xticks(rotation=90)
plt.title("Boxplots for Outliers")
plt.show()
```

# Bivariate Analysis
Correlation Heatmap (Numerical Columns)

```python
num_cols = 2
num_rows = (len(categorical_columns) + 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(18, num_rows * 5))
axes = axes.flatten()
for i, col in enumerate(categorical_columns):
    top_categories = df[col].value_counts().nlargest(10).index

    sns.countplot(x=df[col], palette="viridis", order=top_categories, ax=axes[i], width=0.7)
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=20, ha="right", fontsize=10)
    axes[i].set_title(f"Count Plot for {col} (Top 10)", fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel("Count", fontsize=10)


for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Pairplot for Relationships Between Numerical Variables

```python
sns.pairplot(df[numerical_columns])
plt.show()
```

# Boxplot of Numerical Features vs. Categorical Feature

```python
cat_col = categorical_columns[0]
num_col = numerical_columns[0]

top_categories = df[cat_col].value_counts().index[:10]
df_filtered = df[df[cat_col].isin(top_categories)]

plt.figure(figsize=(14, 6))
sns.boxplot(x=cat_col, y=num_col, data=df_filtered)

plt.xticks(rotation=45, ha='right')
plt.title(f"Boxplot of {num_col} vs. {cat_col} (Top 10 Categories)")
plt.show()
```
